```python
#check for missing values/ inconsistent values
#we use isnull() function
#data cleaning
data.isnull().sum()
```

```python
#create a subset with the columns to use
school_subset = pandas.DataFrame(data,columns=['ids','Major','Math','Height','Gender','Weight','Smoking','English','Reading','Athlete'])
school_subset.head()
```

```python
school_subset.shape
```

```python
school_subset.isnull().sum()
```

```python
# we can clean by dropping the empty cells
# you use dropna function
#school_subset.dropna(inplace=true)
#alternatively
#replace the missing values with a mean of the column
mean_height  = school_subset["Height"].mean()
mean_weight  = school_subset["Weight"].mean()
mean_smoking  = school_subset["Smoking"].mean()
mean_reading  = school_subset["Reading"].mean()
mean_math = school_subset["Math"].mean()
mean_gender  = school_subset["Gender"].mean()
mean_english  = school_subset["English"].mean()
```

```python
#use a function of fillna() to clean the empty cells
school_subset['Height'].fillna(value=mean_height,inplace=True)
school_subset['Weight'].fillna(value=mean_weight,inplace=True)
school_subset['Smoking'].fillna(value=mean_smoking,inplace=True)
school_subset['Reading'].fillna(value=mean_reading,inplace=True)
school_subset['Math'].fillna(value=mean_math,inplace=True)
school_subset['Major'].fillna(value='Philosophy',inplace=True)
school_subset['Gender'].fillna(value='neutral',inplace=True)
school_subset['English'].fillna(value=mean_english,inplace=True)
```
