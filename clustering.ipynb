```python
#clustering is unsupervised machine learning method of identifying and ungrouping similar data
#data is placed into groups based on similarities
#clustering is also knoen as segmentation

import pandas
data = pandas.read_csv('https://modcom.co.ke/datasets/Mall_Customers.csv')
data.head()
```

```python
#dataset cleaning
data.dropna(inplace=True)
#split the dataset into age,income and score
array= data.values
x=array[:,2:5]



#fit data in your model
from sklearn.cluster import KMeans
model=KMeans(n_clusters=10,random_state=42)
model.fit(x)
```

```python
#view the clusters
means =model.cluster_centers_
clusters=pandas.DataFrame(means,columns =('Age','Income','Score'))
clusters
```
