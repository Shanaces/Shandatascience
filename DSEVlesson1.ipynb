```python
import pandas
data = pandas.read_csv("http://modcom.co.ke/datasets/data.csv")
data
```

```python
data.describe()
```

```python
# split to x and y
array = data.values
x = array[:,0:8] #ingredients
y = array[:,8] #type of cake
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
print('model predicted',predictions)
print('expected' , y_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy', accuracy_score(y_test,predictions))
```

```python
recipe = [[17,10,5,6,1,8,1,2]]
outcome = model.predict(recipe)
print('youll be cooking',outcome)
```
