```python
#in this section we are going to analyze bank information
#the data shows details of bank members, such as Gender, loan amount, education, applicant income,loan amount etc.
#we will analyze this data to get some insights about the bank members

import pandas
data = pandas.read_csv('https://modcom.co.ke/datasets/bank.csv')
data
```

```python
#we create graphs using seaborn and matplotlib
#if youre using vs code you have to install pip3 install seaborn and pip3 install matpotlib

import seaborn as sns
import matplotlib.pyplot as plt

plt.style.available
```

```python
plt.style.use('seaborn-v0_8')

#we create a count plot to show proportion of gender
```

```python
sns.countplot(data=data,x='Gender',palette ='Reds')
plt.title('distribution of gender status')
plt.xlabel('Gender')
plt.ylabel('number of members')
```

```python
data.groupby('Property_Area').size().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
data.groupby('Education').size().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
import numpy
sns.barplot(x='Gender',y='LoanAmount',data=data, estimator=numpy.mean, palette = 'magma',hue='Married')
plt.title('distribution of gender status')
plt.xlabel('Gender')
plt.ylabel('loan amount')
```

```python
#swarm plot
colors= ['#3949AB','#2E7D32','#546E7A']
sns.swarmplot(x='Gender',y='ApplicantIncome',data=data,palette='magma',hue='Education')
```

```python
#scatterplot
sns.scatterplot(x='LoanAmount', y='ApplicantIncome',data=data, palette='Reds',hue='Gender')
```
