```python
import pandas
data = pandas.read_csv('https://modcom.co.ke/datasets/Mall_Customers.csv')
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=data,x='Gender',palette ='Greens')
plt.title('Distribution of gender ')
plt.xlabel('Gender')
plt.ylabel('number of customers')
#There are more female customers in the mall compared to male customers
```

```python
data.groupby('Gender').size().plot(kind='pie',autopct='%1.1f%%')
plt.show()
#There are more female customers compared to male customers in the mall
```

```python
import numpy
sns.barplot(x='Gender',y='Annual Income',data=data, estimator=numpy.mean, palette = 'magma')
plt.title('Distribution of gender vs annual income')
plt.xlabel('Gender')
plt.ylabel('Annual Income')
#The male annual income is more than the female annual income
```
